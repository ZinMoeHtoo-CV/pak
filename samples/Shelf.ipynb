# Multiple-View Multiple People Shelf

Dataset: http://campar.in.tum.de/Chair/MultiHumanPose

```python
%matplotlib inline
import json; from pprint import pprint
Settings = json.load(open('settings.txt'))
import matplotlib.pyplot as plt
import numpy as np
import sys
sys.path.insert(0,'../')
#from pak.datasets.EPFL_Campus import EPFL_Campus
from pak.datasets.Shelf import Shelf

root = Settings['data_root']
shelf = Shelf(root)

FRAME = 300
X, Y, Calib = shelf.get_frame(FRAME)

fig = plt.figure(figsize=(16,8))


colors = ['red', 'blue', 'yellow']

for idx, (im, P) in enumerate(zip(X, Calib)):
    ax = fig.add_subplot(2, 3, idx+1)
    ax.imshow(im)
    ax.axis('off')
    
    for pid, person in enumerate(Y):
        if person is None:
            continue
        for x,y,z in person:
            p3d = np.array([x,y,z,1]) 
            u,v,s = P @ p3d
            u = u/s
            v = v/s
            ax.scatter(u, v, color=colors[pid])   

plt.show()
```
