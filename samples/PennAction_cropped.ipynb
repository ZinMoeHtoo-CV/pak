```python
%matplotlib inline
import json; from pprint import pprint
Settings = json.load(open('settings.txt'))
import matplotlib.pyplot as plt
import numpy as np
import sys
sys.path.insert(0,'../')
from pak.datasets.PennAction_cropped import PennAction_cropped

root = Settings['data_root']
penn = PennAction_cropped(root)

folder = penn.labels_folder

vid, nframes = penn.get_random_train_id()
print(nframes)

im, gt = penn.get_frame(vid, 0)



fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(111)
ax.imshow(im)
ax.axis('off')

for x,y,v in gt:
    if v > 0:
        ax.scatter(x, y)


plt.show()
```
