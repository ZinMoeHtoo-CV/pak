# CUHK03
Original paper: https://www.cv-foundation.org/openaccess/content_cvpr_2014/papers/Li_DeepReID_Deep_Filter_2014_CVPR_paper.pdf

The dataset has 3 subsets:
* **labeled**: bounding boxes are labeled by humans
* **detected**: bounding boxes are estimated by a pedestrian detector
* **testsets**: contains the testing protocol

```python
%matplotlib inline
import sys
sys.path.insert(0,'../')
from pak import datasets
from pak import utils
import matplotlib.pyplot as plt


root = '/home/julian/Data/pak_test1'
target_w, target_h = 60, 160

cuhk03 = datasets.cuhk03(root, target_w=target_w, target_h=target_h)


X, Y = cuhk03.get_train_raw('labeled')

n = 6
fig = plt.figure(figsize=(16,8))
for i in range(n):
    fig.add_subplot(1, n, i+1).imshow(X[i+200])
plt.show()
```
